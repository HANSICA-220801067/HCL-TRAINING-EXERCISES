```python
'''Problem 1: You’re given a list of product names with inconsistent casing and extra spaces:
names = ["  alpha Case  ", "Beta   unit", "gAmMa MODULE  "]
Use map with a lambda to strip spaces and convert to Title Case (e.g., "Alpha Case").'''

#solution
names=list(map(str, input().split(",")))
updated_names=list(map(lambda x:x.strip().title(),names))
print(updated_names)
```

```python
'''Create a Price Update Pipeline (map → filter → reduce), give n list of product prices, apply 10% discount,
keep only items ≥ ₹200 after discount and compute the final bill total.'''
#solution
from functools import reduce
prices = list(map(float, input().split()))
discounted = list(map(lambda x: x * 0.9, prices))
new = list(filter(lambda x: x >= 200, discounted))
total = reduce(lambda x, y: x + y,new, 0)
print("Final Bill:", total)
```

```python
#Calculate Weighted Score Aggregationusing reduce when given lists of scores and weights.
#Solution:
from functools import reduce
scores = list(map(float, input().split()))
weights = list(map(float, input().split()))
weightsum = reduce(lambda acc, pair: acc + pair[0] * pair[1],zip(scores, weights),0)
score = weightsum / total if total != 0 else 0
print(score)
```
