```python
import csv
books=[{"title":"B1","author":"ab","price":100},{"title":"B2","author":"cd","price":150}]
f=open("books.csv","w",newline="")
w=csv.DictWriter(f,fieldnames=books[0].keys())
w.writeheader()
w.writerows(books)
f.close()
```
