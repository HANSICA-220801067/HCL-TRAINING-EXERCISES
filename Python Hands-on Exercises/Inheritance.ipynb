```python
class Writer:
    def write(self):
        print("Writing content")
class Reader:
    def read(self):
        print("Reading content")
class Editor(Writer, Reader):
    pass
e = Editor()
e.write()
e.read()
```

```python
class Employee:
    def __init__(self, salary, bonus):
        self._salary = salary
        self.__bonus = bonus
    def get_bonus(self):
        return self.__bonus
class Manager(Employee):
    def show_details(self):
        print("Salary:", self._salary)
        print("Bonus (using method):", self.get_bonus())
m = Manager(50000, 10000)
m.show_details()
```

```python
class Logger:
    def log(self, message):
        print("Log:", message)
class FileLog(Logger):
    def log(self, message):
        super().log(message)
        print("Save:", message)
f = FileLog()
f.log("Start")
```

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
    def start(self):
        print(self.brand,"start")
class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model
    def drive(self):
        print(self.brand, self.model,"drive")
class ElectricCar(Car):
    def __init__(self, brand, model, battery):
        super().__init__(brand, model)
        self.battery = battery
    def charge(self):
        print(self.brand, self.model, "is charging", self.battery, "kWh battery.")
e = ElectricCar("Tesla", "Model 3", 75)
e.start()
e.drive()
e.charge()
```
