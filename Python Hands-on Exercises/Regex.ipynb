```python
import re
sentence =input()
keyword =input()
if re.search(keyword, sentence):
    print("found")
else:
    print("Not found")
```

```python
import re
emails = input()
domains = re.findall(r'@(\w+\.\w+)', emails)
print(domains)
```

```python
import re
card = input()
masked = re.sub(r'\d(?=\d{4})', '*', card)
print(masked)
```

```python
import re
text = input()
if re.match(r'\d{4}-\d{2}-\d{2}', text):
    print("Starts")
else:
    print("does not start")
```
