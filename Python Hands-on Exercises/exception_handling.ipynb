```python
try:
    a=int(input())
    b=int(input())
    print(a/b)
except Exception as e:
    open("error.txt","a").write(str(e)+"\n")
```

```python
class InvalidEmailError(Exception):pass
email=input()
if "@" not in email:
    raise InvalidEmailError("Invalid Email")
print("Valid")
```

```python
li=list(map(int,input().split()))
for i in li:
    try:
        print(100/i)
    except ZeroDivisionError:
        print("zero division")
```

```python
import json
try:
    print(json.load(open(input())))
except FileNotFoundError:
    print("File not found")
except json.JSONDecodeError:
    print("Invalid JSON")
```

```python
def check(a):
    if a<18:
      raise ValueError("Underage")
    else:
      print("Eligible")
check(int(input()))
```

```python
try:
    a=int(input())
    b=int(input())
    print(a/b)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
```
