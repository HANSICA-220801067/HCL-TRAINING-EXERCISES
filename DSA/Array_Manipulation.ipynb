```python
#dutch problem
arr=[2,1,2,0,1,0,0,2,0,0,1]
c1=c2=c0=0
for i in arr:
  if i==0:
    c0+=1
  elif i==1:
    c1+=1
  else:
    c2+=1
print([0]*c0+[1]*c1+[2]*c2)
```

```python
#traversal
arr=[1,4,7,3,7,9]
for i in range(len(arr)):
  print(arr[i])
```

```python
#Insert Element at End
arr=[10,20,30,40]
n=len(arr) #4
new=[0]*(n+1)
for i  in range(n):
  new[i]=arr[i]
new[n]=50
print(new)

new.append(60)
print(new)
```

```python
#Insert Element at Specific Position
arr=[10,20,30,40,50]
n=len(arr)
new=[0]*(n+1)
pos=2
val=25
for i in range(pos):
  new[i]=arr[i]
new[pos]=val
for i in range(pos,n):
  new[i+1]=arr[i]
print(new)

new.insert(3,27)
print(new)
```

```python
#Delete an element at specific position
arr=[10,20,30,40,50]
new=[0]*(n-1)
pos=2
for i in range(pos):
  new[i]=arr[i]
for i in range(pos+1,n):
  new[i-1]=arr[i]
print(new)
```

```python
#reverse an array two pointer
arr=[10,20,30,40,50]
n=len(arr)
l=0
r=n-1
while l<r:
  temp=arr[l]
  arr[l]=arr[r]
  arr[r]=temp
  l=l+1
  r=r-1
print(arr)
```

```python
#linear search
arr=[1,2,3,4,5,6,7]
s=5
for i in range(len(arr)):
  if arr[i]==s:
    print("value found at index = ",i)
```

```python
#Find Maximum Element
arr=[1,2,3,5,6,8,3,6]
max_val=arr[0]
for i in range(1, len(arr)):
  if arr[i]>max_val:
    max_val=arr[i]
print(max_val)
```

```python
#Find Minimum Element
arr=[1,2,3,5,6,8,3,6]
min_val=arr[0]
for i in range(1, len(arr)):
  if arr[i]<min_val:
    min_val=arr[i]
print(min_val)
```

```python
#Array Rotation (Left Rotation)
arr=[1,2,3,4,5]
temp=arr[0]
for i in range(1,len(arr)):
  arr[i-1]=arr[i]
arr[len(arr)-1] = temp
print(arr)
```

```python
#right rotation
arr=[1,2,3,4,5]
temp=arr[len(arr)-1]
for i in range(len(arr)-1,0,-1):
  arr[i]=arr[i-1]
arr[0]=temp
print(arr)
```

```python
#left rotation by k steps
arr=[1,2,3,4,5]
k=7
k=k%len(arr)
for j in range(k):
  temp=arr[0]
  for i in range(1,len(arr)):
    arr[i-1]=arr[i]
  arr[len(arr)-1]=temp
print(arr)
```

```python
#right rotation by k steps
arr=[1,2,3,4,5]
k=7
k=k%len(arr)
for j in range(k):
  temp=arr[len(arr)-1]
  for i in range(len(arr)-1,0,-1):
    arr[i]=arr[i-1]
  arr[0]=temp
print(arr)
```

```python
#sub array
arr=[1,2,3,4,5,6,7]
n=len(arr)
for i in range(n):
  for j in range(i,n):
    for k in range(i,j+1):
      print(arr[k],end=" ")
    print()
```

```python
#two sum logic
arr=[1,2,3,4,5,6,7]
t=9
n=len(arr)
for i in range(n):
  for j in range(i+1,n):
    if arr[i]+arr[j]==t:
      print(i,j)
```

```python
#Frequency Counting
arr=[1,2,3,5,1,3,1,2,3,5,8,35,3]
freq={}
for i in arr:
  if i in freq:
    freq[i]+=1
  else:
    freq[i]=1
print(freq)
```
