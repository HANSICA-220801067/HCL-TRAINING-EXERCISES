```python
#easy : Maximum amount of water a container can store
n=int(input())
h=list(map(int, input().split()))
l,r=0,n-1
area=0
while l<r:
    a=min(h[l], h[r]) * (r - l)
    area=max(area,a)
    if h[l]<h[r]:
        l=l+1
    else:
        r=r-1
print(area)
```

```python
#2 : excel column number
n=int(input())
r=""
while n>0:
    n=n-1
    r=chr(n%26+ord('A'))+r
    n//=26
print(r)
```

```python
#3. climbing stairs (like fibonanci series )
n=int(input())
if n==1:
    print(1)
elif n==2:
    print(2)
else:
    a,b=1,2
    for i in range(3,n+1):
        a,b=b,a+b
    print(b)
```
